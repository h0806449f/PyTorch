## R-CNN
缺點 : 多個隨機性搜索的區域, 可能重疊, 且各個區域都獨立通過CNN -> 速度非常慢  
r-cnn.jpg

## Fast-R-CNN
改進 : 1) 一次對整張圖像卷積 2) 隨機搜索不同區域 -> 但縮放為相同大小 3) 一起通過全聯階層  
缺點 : 仍有隨機搜索

fast-rcnn.jpg

## Faster-R-CNN  
改進 : 將隨機搜索改區域提議網格
